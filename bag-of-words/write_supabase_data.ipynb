```python
import os
from dotenv import load_dotenv
from supabase import create_client, Client
import pandas as pd
import json

load_dotenv()
```

```python
# initialize supabase
url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)
```

```python
# read in csv

df = pd.read_csv('grouped_data/data.csv')

df.head(3)
```

```python
# remove unnecessary columns
only_topics_df = df[['id', 'title', 'group']]
only_topics_df = only_topics_df.rename(columns={'title': "name", 'group': "topic"})

only_topics_df.head()
```

```python
# convert to json
json_str = only_topics_df.to_json(orient="records")
json_data = json.loads(json_str)

print(json_data)
```

```python
# update supabase
val = supabase.table('Data').upsert(json_data).execute()

print(val)
```
